```python
import numpy as np
```

```python
if __name__ == "__main__":
    # Load data
    if ('trainingFaces' not in globals()):  # In ipython, use "run -i homework2_template.py" to avoid re-loading of data
        train_images = np.load("mnist_train_images.npy")
        train_labels = np.load("mnist_train_labels.npy")
        test_images = np.load("mnist_test_images.npy")
        test_labels = np.load("mnist_test_labels.npy")
```

```python
def J(true_labels, softmax_matrix):
    return - np.mean(np.sum(true_labels * np.log(softmax_matrix + 1e-5) ,axis = 1))
```

```python
def softmax(weight, train_images):
    def trans(row):
        row_sum = np.sum(np.exp(row))
        return [1.0 * np.exp(ele) / row_sum for ele in row]
    train_images = train_images.dot(weight)
    train_images = np.apply_along_axis(trans, 1, train_images)
    return train_images
```

```python
def grad(true_labels, softmax_matrix, train_images):
    return train_images.T.dot(softmax_matrix - true_labels)
```

```python
def get_accu(train_labels, softmax_matrix):
    train_digit = np.argmax(train_labels, axis = 1)
    predict_digit = np.argmax(softmax_matrix, axis = 1)
    return np.sum(train_digit == predict_digit) * 1.0 / train_labels.shape[0]
```

```python
def gradientDescent (train_images, train_labels, test_images, test_labels, k):
    w = np.zeros((train_images.shape[1], k))  # Or set to random vector
    tol = 1e-8
    lr = 1e-6
    cost = J (train_labels, softmax(w, train_images))
    costNew = -1
    i = 0
    while np.sum(np.square(cost - costNew)) > tol:
        if costNew != -1:
            cost = costNew
        wNew = w - lr * grad(train_labels, softmax(w, train_images), train_images)
        softmax_matrix = softmax(wNew, train_images)
        costNew = J (train_labels, softmax_matrix)
        accuracy = get_accu(train_labels, softmax_matrix)
        print str(i) + ': The training cost is ' + str(costNew)
        print str(i) + ': The accuracy is ' + str(accuracy)
        w = wNew
        i += 1
    return wNew
```

```python
def reportCosts (w, train_images, train_labels, test_images, test_labels, k):
    print "Training cost: {}".format(J(train_labels, softmax(w, train_images)))
    print "Training accuracy: {}".format(get_accu(train_labels, softmax(w, train_images)))
    print "Testing cost:  {}".format(J(test_labels,  softmax(w, test_images)))
    print "Testing accuracy:  {}".format(get_accu(test_labels,  softmax(w, test_images)))
```

```python
w = gradientDescent (train_images, train_labels, test_images, test_labels, k = 10)
reportCosts (w, train_images, train_labels, test_images, test_labels, k = 10)
```
